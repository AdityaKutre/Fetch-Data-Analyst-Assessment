```python
#Checking the Data Quality

# Importing the required libraries
import pandas as pd
import numpy as np

#Reading the dataset
df_users=pd.read_csv('/content/USER_TAKEHOME.csv')
df_transaction=pd.read_csv('/content/TRANSACTION_TAKEHOME.csv')
df_products=pd.read_csv('/content/PRODUCTS_TAKEHOME.csv')
```

```python

# Step 1
#Checking if there are any null present
print('Null\'s in  Users Column')
print(df_users.isnull().sum())
print()
print()
print('Null\'s in  Transaction Column')
print(df_transaction.isnull().sum())
print()
print()
print('Null\'s in  Products Column')
print(df_products.isnull().sum())
```

```python
# Step 2
#Checking if there are duplicate present
print('Duplicates\'s in  Users Column')
print(df_users.duplicated().sum())
print()
print()
print('Duplicates\'s in  Transaction Column')
print(df_transaction.duplicated().sum())
print()
print()
print('Duplicates\'s in  Products Column')
print(df_products.duplicated().sum())
```

```python
#Analysing the data through the graph/visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Set visualization style
sns.set(style="whitegrid")

# Visualization for PRODUCTS_TAKEHOME.csv
plt.figure(figsize=(12, 6))
products_missing = df_products.isnull().sum()
products_missing = products_missing[products_missing > 0]
sns.barplot(x=products_missing.index, y=products_missing.values, palette="viridis")
plt.title('Missing Values in PRODUCTS_TAKEHOME.csv')
plt.ylabel('Count of Missing Values')
plt.xlabel('Columns')
plt.xticks(rotation=45)
plt.show()

# Visualization for TRANSACTION_TAKEHOME.csv
plt.figure(figsize=(12, 6))
transactions_missing = df_transaction.isnull().sum()
transactions_missing = transactions_missing[transactions_missing > 0]
sns.barplot(x=transactions_missing.index, y=transactions_missing.values, palette="magma")
plt.title('Missing Values in TRANSACTION_TAKEHOME.csv')
plt.ylabel('Count of Missing Values')
plt.xlabel('Columns')
plt.xticks(rotation=45)
plt.show()

# Visualization for USER_TAKEHOME.csv
plt.figure(figsize=(12, 6))
users_missing = df_users.isnull().sum()
users_missing = users_missing[users_missing > 0]
sns.barplot(x=users_missing.index, y=users_missing.values, palette="coolwarm")
plt.title('Missing Values in USER_TAKEHOME.csv')
plt.ylabel('Count of Missing Values')
plt.xlabel('Columns')
plt.xticks(rotation=45)
plt.show()

```

```python
# Distribution plot for PRODUCTS_TAKEHOME.csv (BARCODE field)
plt.figure(figsize=(10, 6))
sns.histplot(df_products['BARCODE'].dropna(), bins=50, kde=True, color='blue')
plt.title('Distribution of BARCODE in PRODUCTS_TAKEHOME.csv')
plt.xlabel('Barcode')
plt.ylabel('Frequency')
plt.show()


# Distribution plot for USER_TAKEHOME.csv (BIRTH_DATE field)
df_users['BIRTH_DATE'] = pd.to_datetime(df_users['BIRTH_DATE'], errors='coerce')
plt.figure(figsize=(10, 6))
sns.histplot(df_users['BIRTH_DATE'].dt.year.dropna(), bins=50, kde=True, color='purple')
plt.title('Distribution of Birth Years in USER_TAKEHOME.csv')
plt.xlabel('Year of Birth')
plt.ylabel('Frequency')
plt.show()

# Distribution plot for TRANSACTION_TAKEHOME.csv (FINAL_SALE field)
df_transaction['FINAL_SALE'] = pd.to_numeric(df_transaction['FINAL_SALE'], errors='coerce')
plt.figure(figsize=(10, 6))
sns.histplot(df_transaction['FINAL_SALE'].dropna(), bins=50, kde=True, color='green')
plt.title('Distribution of FINAL_SALE in TRANSACTION_TAKEHOME.csv')
plt.xlabel('Final Sale Amount')
plt.ylabel('Frequency')
plt.show()
```
